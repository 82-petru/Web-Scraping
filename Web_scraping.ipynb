```python
import requests
from bs4 import BeautifulSoup

prefix = "https://content.codecademy.com/courses/beautifulsoup/"
webpage_response = requests.get('https://content.codecademy.com/courses/beautifulsoup/shellter.html')

webpage = webpage_response.content
soup = BeautifulSoup(webpage, "html.parser")
#print(soup)
turtle_links = soup.find_all("a")
# print(turtle_links)
links = []
#go through all of the a tags and get the links associated with them:
for a in turtle_links:
  links.append(prefix+a["href"])

#print(links[0])
webpage = requests.get(links[0])
turtle = BeautifulSoup(webpage.content, "html.parser")
#print(turtle)
turtle_stats = turtle.select("ul")
# print(turtle_stats)
turtle_name = turtle.select('.name')
#print(turtle_name)
#Define turtle_data:
turtle_data = {}
#follow each link:
for link in links:
  webpage = requests.get(link)
  turtle = BeautifulSoup(webpage.content, "html.parser")
  turtle_name = turtle.select('.name')[0].get_text()
  turtle_stats =turtle.select('ul')[0].get_text('|').split('|')
  #Add your code here:
  turtle_data[turtle_name] = turtle_stats
print(turtle_data)  
```

**Chocolate Scraping with Beautiful Soup**

**Import the libraries**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Let’s make a request to this site to get the raw HTML. 
Create a BeautifulSoup object called soup to traverse this HTML.
If you want, print out the soup object to explore the HTML.

```python
url = "https://content.codecademy.com/courses/beautifulsoup/cacao/index.html"
webpage_response = requests.get(url)
webpage = webpage_response.content
soup = BeautifulSoup(webpage, "html.parser")
```

Put all of the ratings into a list. 
Use a command on the soup object to get all of the tags that contain the ratings.

**How are ratings distributed?**

```python
ratings_tags = soup.select('.Rating')
```

Create an empty list called ratings to store all the ratings in. Loop through the ratings tags and get the text contained in each one. Add it to the ratings list. As you do this, convert the rating to a float, so that the ratings list will be numerical. This should help with calculations later.

```python
ratings = []
for rating in ratings_tags:
  tag_all = rating.get_text()
  ratings.append(tag_all)
del ratings[0]
tag_numeric = []
for item in ratings:
  tag_numeric.append(float(item))
```

Using Matplotlib, create a histogram of the ratings values:

```python
plt.hist(tag_numeric, edgecolor = 'black')
plt.show()
```

We want to now find the 10 most highly rated chocolatiers. One way to do this is to make a DataFrame that has the chocolate companies in one column, and the ratings in another. Then, we can do a groupby to find the ones with the highest average rating.

```python
companies_tags = soup.find_all(attrs = {'class' : 'Company'})
```

Just like we did with ratings, we now want to make an empty list to hold company names. Loop through the tags containing company names, and add the text from each tag to the list you just created.

```python
companies = []
for company in companies_tags[1:]: 
    companies.append(company.get_text())
```

Create a DataFrame with a column “Company” corresponding to your companies list, and a column “Ratings” corresponding to your ratings list.

```python
# one way to create dataframe
d = {'Company' : companies, 'Rating' : tag_numeric}
df = pd.DataFrame(d)
df.head()
```

```python
# second way to create a dataframe 
d = d = {'Company' : companies, 'Rating' : tag_numeric}
df = pd.DataFrame.from_dict(d)
# df.head()
```

Use .groupby to group your DataFrame by Company and take the average of the grouped ratings.

```python
ratings_average = df.groupby('Company').Rating.mean()
print(ratings_average)
```

Then, use the .nlargest command to get the 10 highest rated chocolate companies. Print them out.

```python
ten_highest = ratings_average.nlargest(10)
ten_highest
```

**Check the chocolate cocoa level**

We want to see if the chocolate experts tend to rate chocolate bars with higher levels of cacao to be better than those with lower levels of cacao.
Using the same methods you used in the last couple of tasks, create a list that contains all of the cocoa percentages. Store each percent as an integer, after stripping off the % character.

```python
# Use a command on the soup object to get all of the tags that contain the cacoa percentage.
cocoa_percent_tags = soup.find_all(attrs = {'class' : "CocoaPercent"})
cocoa_percent = []
for cocoa in cocoa_percent_tags[1:]:
    remove_trailing = cocoa.get_text().strip('%')
    cocoa_percent.append(float(remove_trailing))    
```

Using Matplotlib, create a histogram of the cocoa percentege values:

```python
plt.hist(cocoa_percent, edgecolor = 'black')
plt.show()
```

Add the cocoa percentages as a column called "CocoaPercentage" in the DataFrame that has companies and ratings in it.

```python
df["CocoaPercentage"] = cocoa_percent 
df.head()
```

Make a scatterplot of ratings (your_df.Rating) vs percentage of cocoa (your_df.CocoaPercentage).

```python
plt.scatter(df.CocoaPercentage, df.Rating, color = 'red')
z = np.polyfit(df.CocoaPercentage, df.Rating, 1)
line_function = np.poly1d(z)
plt.plot(df.CocoaPercentage, line_function(df.CocoaPercentage), "r--")
plt.show()
```
